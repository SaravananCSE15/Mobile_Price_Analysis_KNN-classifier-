Taking "Price_range" as Dependent variable

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


x2=mobile.drop('price_range', axis=1)
y2 = mobile['price_range']
```

Apply KNeighborsClassifier to Predict Performance

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(x2, y2, test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
!pip install eli5
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(knn, random_state=1).fit(x2, y2)
importances = perm.feature_importances_

features = x2.columns

for i, feature in enumerate(features):
    print(f"{feature}: {importances[i]}")
```

XBGRegressor

```python
import xgboost as xgb

#create XGboost Regressor
xgboost_model=xgb.XGBRegressor()

#train the model
xgboost_model.fit(X_train,y_train)


from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

#make Predictions
xgboost_predictions = xgboost_model.predict(X_test)
mse=mean_squared_error(y_test,xgboost_predictions)
mae=mean_absolute_error(y_test,xgboost_predictions)
r2=r2_score(y_test,xgboost_predictions)

print("score",r2)
```

```python
from catboost import CatBoostClassifier

# Create a CatBoostClassifier object
model = CatBoostClassifier()

# Fit the model to the training data
model.fit(X_train, y_train)

# Predict the labels of the test data
y_pred = model.predict(X_test)

# Evaluate the model accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.ensemble import AdaBoostClassifier

# Create an AdaBoostClassifier object
abc = AdaBoostClassifier(n_estimators=50, learning_rate=1)

# Train the model
abc.fit(X_train, y_train)

# Make predictions
y_pred_abc = abc.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_abc)
print("Accuracy:", accuracy)

# Get feature importances
importances = abc.feature_importances_

# Print feature importances
for i, feature in enumerate(features):
    print(f"{feature}: {importances[i]}")
```

```python
from sklearn.ensemble import GradientBoostingClassifier

# Create a GradientBoostingClassifier object
gbc = GradientBoostingClassifier()

# Train the model
gbc.fit(X_train, y_train)

# Make predictions
y_pred_gbc = gbc.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred_gbc)
print("Accuracy:", accuracy)

# Get feature importances
importances = gbc.feature_importances_

# Print feature importances
for i, feature in enumerate(features):
    print(f"{feature}: {importances[i]}")
```
